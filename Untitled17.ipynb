```python
#House price prediction using linear regression
#import nessescary library
#%pip install msvc-runtime
%pip install seaborn
%pip install scikit-learn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sess
import sklearn
```

```python
data=pd.read_csv("/content/drive/MyDrive/Housing.csv")
data.head()
```

```python
#Dataset Structure
data.shape
data.describe()
data.info()
```

```python
#Data cleaning
data.isnull().sum()
```

```python
#data visualisation
figure,axes=plt.subplots(2,2,figsize=(10,5))
plt1=sess.boxplot(data.area, ax=axes[0,0])
plt2=sess.boxplot(data['price'],ax=axes[0,1])
plt3=sess.boxplot(data['bedrooms'],ax=axes[1,0])
plt4=sess.boxplot(data['bathrooms'],ax=axes[1,1])
plt.tight_layout()
```

```python
#there is outliers in price and area
#treatment for outliers in price and area
plt.boxplot(data.price)
q1=data.price.quantile(0.25)
q2=data.price.quantile(0.75)
iqr=q2-q1
data=data[(data.price >= q1 - 1.5*iqr) & ( data.price <= q2 + 1.5*iqr)]
plt.show()
```

```python
# treatment for outlier in area
plt.boxplot(data.area)
q1=data.area.quantile(0.25)
q2=data.area.quantile(0.75)
iqr=q2-q1
data=data[(data.area >= q1 - 1.5*iqr) & (data.area <= q2 + 1.5*iqr )]
plt.show()
```

```python
figure,axes = plt.subplots(2,2,figsize=(10,5))
plt5=sess.boxplot(data.price,ax=axes[0,0])
plt6=sess.boxplot(data.area, ax=axes[0,1])
plt.show()
```

```python
#split the dataset
from sklearn.model_selection import train_test_split
np.random.seed(0)
data=data[['price','area','bedrooms','bathrooms']]
train,test=train_test_split(data,train_size=0.7,test_size=0.3,random_state=100)
```

```python
#Rescale the features
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
col=['price','area','bedrooms','bathrooms']
train[col]=scalar.fit_transform(train[col])
train.head()
```

```python
#Model building
y_train=train.pop('price')
x_train=train
from  sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x_train,y_train)
print(train.columns)
train.head()
print(lm.intercept_)
```

```python
y_train_price=lm.predict(x_train)
res=y_train-y_train_price
sess.distplot(res)
```

```python
plt.scatter(y_train,res)
plt.show()
```

```python
test['price']=data['price']
vars=['price','area','bedrooms','bathrooms']
test[vars]=scalar.fit_transform(test[vars])
test.head()
```

```python
test['price']=data['price']
y_test=test.pop('price')
x_test=test
y_pred=lm.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
fig = plt.figure()
plt.scatter(y_test,y_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)
plt.xlabel('y_test', fontsize=18)
plt.ylabel('y_pred', fontsize=16)
```
